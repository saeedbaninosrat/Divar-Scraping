```python
import pandas as pd
import re
df = pd.read_csv(r'./19sep.csv', encoding='utf8')
df = pd.DataFrame(df)
```

```python

# print(df.test2)

# df = df['test2']

# print(len(df.index))
show = []
motor = []
chassis = []
fchassis = []
bchassis = []
bodys = []
insurance = []
gearbox = []
price = []
distance = []
year = []
color = []

for i in range(31708):
    dyc = re.findall(r"(.*), (.*), (.*)", df['Distance Year Color'][i])
    distance.append(dyc[0][0])
    year.append(dyc[0][1])
    color.append(dyc[0][2])
    show.append(re.findall(r"نمایشگاه,(.*?),", df['test2'][i]))
    motor.append(re.findall(r",وضعیت موتور,(.*?),", df['test2'][i]))
    chassis.append(re.findall(r",وضعیت شاسی‌ها,(.*?),", df['test2'][i]))
    insurance.append(re.findall(
        r",مهلت بیمهٔ شخص ثالث,(.*?),", df['test2'][i]))
    fchassis.append(re.findall(r",شاسی جلو,(.*?),", df['test2'][i]))
    bchassis.append(re.findall(r",شاسی عقب,(.*?),", df['test2'][i]))
    bodys.append(re.findall(r",وضعیت بدنه,(.*?),", df['test2'][i]))
    gearbox.append(re.findall(r",گیربکس,(.*?),", df['test2'][i]))
    price.append(re.findall(r",قیمت فروش نقدی,(.*) تومان", df['test2'][i]))

    # print(df['test2'][i])
    # i +=1

df['نمایشگاه'] = show
df['وضعیت موتور'] = motor
df['وضعیت شاسی‌ها'] = chassis
df['شاسی جلو'] = fchassis
df['شاسی عقب'] = bchassis
df['وضعیت بدنه'] = bodys
df['مهلت بیمهٔ شخص ثالث'] = insurance
df['گیربکس'] = gearbox
df['قیمت فروش نقدی'] = price
df['کارکرد'] = distance
df['سال تولید'] = year
df['رنگ'] = color

print(df)
# type(df)
```

```python
df = df.drop('Distance Year Color', axis=1)
df = df.drop('test2', axis=1)
print(df)
```
